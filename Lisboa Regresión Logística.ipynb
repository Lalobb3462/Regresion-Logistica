**instant_bookable vs availability_30, host_acceptance_rate, price & beds**

```python
#Declaramos las variables dependientes e independientes para la regresión logística
Vars_Indep=df[['availability_30','host_acceptance_rate','price','beds']] #De tipo numérico
Var_Dep=df['instant_bookable']
```

```python
#Redefinimos las variables
X= Vars_Indep
y= Var_Dep
```

```python
#Dividimos el conjunto de datos en la parte de entrenamiento y prueba:
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=None)
```

```python
#Se escalan todos los datos
escalar = StandardScaler()
```

```python
#Para realizar el escalamiento de las variables "X" tanto de entrenamiento como de prueba, utilizaremos
X_train = escalar.fit_transform(X_train)
X_test = escalar.transform(X_test)
```

```python
#Definimos el algoritmo a utilizar
from sklearn.linear_model import LogisticRegression
algoritmo = LogisticRegression()
```

```python
#Entrenamos el modelo
algoritmo.fit(X_train, y_train)
```

```python
#Realizamos una predicción
y_pred = algoritmo.predict(X_test)
y_pred
```

```python
#Verifico la matriz de Confusión
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print ('Matriz de Confusión: ')
print(matriz)
```

```python
#Calculo la precisión del modelo
from sklearn.metrics import precision_score

precision1 = precision_score(y_test, y_pred, average="binary", pos_label="t")
print("Precisión del modelo: ")
print(precision1)
```

```python
from sklearn.metrics import precision_score

precision2 = precision_score(y_test, y_pred, average="binary", pos_label="f")
print("Precisión del modelo: ")
print(precision2)
```

```python
#Calculo la exactitud del modelo
from sklearn.metrics import accuracy_score

exactitud1 = accuracy_score(y_test,y_pred)
print("Exactitud del modelo: ")
print(exactitud1)
```

```python
#Calculo la sensibilidad del modelo
from sklearn.metrics import recall_score

sensibilidad1 = recall_score(y_test, y_pred, average="binary", pos_label="t")
print("Sensibilidad del modelo: ")
print(sensibilidad1)
```

```python
from sklearn.metrics import recall_score

sensibilidad2 = recall_score(y_test, y_pred, average="binary", pos_label="f")
print("Sensibilidad del modelo: ")
print(sensibilidad2)
```
